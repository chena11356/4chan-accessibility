```python
from matplotlib import pyplot as plt
import numpy as np
import json
import os
from io import StringIO
from html.parser import HTMLParser

class MLStripper(HTMLParser):
    def __init__(self):
        super().__init__()
        self.reset()
        self.strict = False
        self.convert_charrefs= True
        self.text = StringIO()
    def handle_data(self, d):
        self.text.write(d)
    def get_data(self):
        return self.text.getvalue()

def strip_tags(html):
    s = MLStripper()
    s.feed(html)
    return s.get_data()

def indexBoardById(boardjson):
    boarddict = {}
    for th in boardjson:
        posts = th['posts']
        postid = posts[0]['no']
        boarddict[postid] = th
    return boarddict

def loadBoard(boardname,month,day,year):
    folder = f"../data/{boardname}/"
    match = [filename for filename in os.listdir(folder) 
                if filename.startswith(f"{boardname}Data{year}{month}{day}")]
    if len(match) == 0:
        print("no matching files")
        return
    with open(folder+match[0], 'r') as j:
        contents = json.loads(j.read())
        return indexBoardById(contents)
    
def getThreadId(thread):
    return thread['posts'][0]['no']

def returnTopicLine(thread):
  return strip_tags(thread['posts'][0]['sub'])+'\n'+strip_tags(thread['posts'][0]['com'])

def processSentiment(s):
    if s['label'] == "NEGATIVE":
      return round(100-s['score']*100)
    return round(s['score']*100)

def generateSentHistForThread(thread,nbins=3):
  a = [processSentiment(sentiment_pipeline([strip_tags(p['com'])])[0])
     for p in thread['posts']]
  fig, ax = plt.subplots(figsize =(10, 7))
  ax.hist(a, bins = [100/nbins*i for i in range(nbins+1)])
  plt.xlabel('Positivity buckets')
  plt.xlabel('Number of Comments')
  plt.show()

def displayThreadInfo(board,id,nbins=4):
  thread = getThread(board,id)
  print(returnTopicLine(thread))
  generateSentHistForThread(thread,nbins=nbins)

def displayThreadInfoForMultiple(board,nTimes=10,skipFirst=True):
    i = 0
    for thread in b.values():
        if skipFirst:
            skipFirst = False
            continue
        if i < nTimes:
            try:
              displayThreadInfo(b, getThreadId(thread))
              i+=1
            except Exception as e:
              print(e)
        else:
            break
```
